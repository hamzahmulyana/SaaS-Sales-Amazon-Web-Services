- **Distribution for Numeric Data**

```python
#Create an empty list to store data
dist = []

#Looping for each column in df_num
for col in df_num.columns:
    #Perform normality test with Shapiro-Wilk
    stat, p_value = shapiro(df[col])
    
    #Determining distribution based on p-value
    distribution = "Normal Distribution" if p_value > 0.05 else "Not-Normal Distribution"
    
    #Save the results to a list
    dist.append([col, stat, p_value, distribution])

#Create a DataFrame from a list
df_dist = pd.DataFrame(dist, columns=['Column', 'Statistic', 'P-Value', 'Distribution'])
df_dist
```

**Interpretation:**

The table shows the results of statistical tests for several data: Sales, Quantity, Discount, and Profit. The "P-Value" column provides the results of the significance test to test the hypothesis.

P-Value = 0: In this context, a p-value of 0 (or very close to 0) indicates very strong evidence to reject the null hypothesis. This means that we can conclude with great confidence that the data in this column is not normally distributed.

```python
#Create an empty list to store data
outliers_info = []

for col in df_num.columns:
    #Calculated Q1, Q2, dan IQR
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    #Calculates upper and lower limits
    batas_bawah = Q1 - 1.5 * IQR
    batas_atas = Q3 + 1.5 * IQR
    
    #Identify outliers
    outliers_condition = (df[col] < batas_bawah) | (df[col] > batas_atas)
    outliers = df[outliers_condition]

    #Added iteration results to outliers_info
    outliers_info.append([col, round(batas_bawah, 2), round(batas_atas, 2), outliers_condition.sum()])

#Turning outliers_info into a dataframe
df_outliers = pd.DataFrame(outliers_info, columns=['Columns', 'Lower Bound', 'Upper Bound', 'Number of Outliers']).set_index('Columns')
df_outliers
```
